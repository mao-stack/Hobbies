```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
import seaborn as sns
```

# Create data

```python
from sklearn.datasets import make_moons

data = make_moons(n_samples=1000, noise=0.2, shuffle=True, random_state=813)

X = pd.DataFrame(data[0])
y = pd.Series(data[1])
```

```python
X.columns = [f"F{i+1}" for i in range(X.shape[1])]
X
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X["F1"][y==0], X["F2"][y==0], c="red")
plt.scatter(X["F1"][y==1], X["F2"][y==1], c="teal")
plt.show()
```

```python
sns.countplot(y)
plt.show()
```

# Modeling

# SVM

```python
from sklearn.svm import SVC

lin = SVC(kernel="linear")
pol3 = SVC(kernel="poly", degree=3)
pol4 = SVC(kernel="poly", degree=4)
pol5 = SVC(kernel="poly", degree=5)
pol6 = SVC(kernel="poly", degree=6)
rbf = SVC(kernel="rbf")
sig = SVC(kernel="sigmoid")

lin.fit(X, y)
pol3.fit(X, y)
pol4.fit(X, y)
pol5.fit(X, y)
pol6.fit(X, y)
rbf.fit(X, y)
sig.fit(X, y)
```

```python
# æ ¼å­çŠ¶ã®ç‚¹ã‚’ç”¨æ„
sample = np.array([(a, b) for a in np.arange(-1.5, 2.7, 0.005) for b in np.arange(-1.5, 1.5, 0.005)])
```

```python
pred=lin.predict(sample)

plt.figure(figsize=(8, 6))

# å…ƒãƒ‡ãƒ¼ã‚¿ã‚’ãƒ—ãƒ­ãƒƒãƒˆ
plt.scatter(X["F1"][y==0], X["F2"][y==0], c="red", s=20)
plt.scatter(X["F1"][y==1], X["F2"][y==1], c="teal", s=20)

# æ ¼å­çŠ¶ã®ç‚¹ã«å¯¾ã—ã¦æŽ¨å®šã—ãŸãƒ‡ãƒ¼ã‚¿ã‚’ãƒ—ãƒ­ãƒƒãƒˆ
plt.scatter(sample[pred==0, 0], sample[pred==0, 1], c="red", s=1, alpha=0.03)
plt.scatter(sample[pred==1, 0], sample[pred==1, 1], c="teal", s=1, alpha=0.03)
plt.title("kernel = linear", fontsize=20)
plt.show()
```
